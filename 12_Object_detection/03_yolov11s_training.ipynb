```python
from pathlib import Path

DATASET_ROOT = Path("/kaggle/working/final-dataset")
DATA_YAML = DATASET_ROOT / "data.yaml"

MODEL_PATH = "/kaggle/working/yolo11s.pt"

print("Dataset exists:", DATASET_ROOT.exists())
print("YAML exists:", DATA_YAML.exists())
print("Model exists:", Path(MODEL_PATH).exists())
```

```python
from ultralytics import YOLO
```

```python
model = YOLO(MODEL_PATH)
```

```python

history = model.train(
    data=str(DATA_YAML),
    epochs=100,
    imgsz=1024,
    batch=16,
    device=0,
    workers=2,
    close_mosaic=10,
    patience=20,
    # Precision & speed
    amp=True,
    lr0=0.005,

    # Augmentations (football-safe)
    mosaic=0.6,
    mixup=0.0,
    copy_paste=0.0,
    freeze=10,

    # Logging
    project="football_yolo11",
    name="yolo11s_1024px",
    exist_ok=True
)
```
