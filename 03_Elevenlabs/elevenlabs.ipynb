```python
import os
import requests
from dotenv import load_dotenv
from elevenlabs.play import play
from elevenlabs.client import ElevenLabs
```

```python
load_dotenv()

ELEVENLABS_API_KEY = os.environ["ELEVENLABS_API_KEY"]
elevenlabs  = ElevenLabs(api_key=ELEVENLABS_API_KEY)
```

```python
prompt = """
    [surprised] Ohh is that so	
    [confident][giggles] You're wrong on this one.	
    [excitedly] Congrats, you've been shortlisted for the next round.	
    [trying to confuse] Are you sure about that?	
    [no emotions] Tell me something about yourself	
    [trying to cheer] No worries, let's move onto the next question.	
    [sadness] Im sorry to inform that you've not been shortlisted.	
"""

model = 'eleven_v3'

audio_stream = elevenlabs.text_to_speech.convert(
    text=prompt,
    voice_id="21m00Tcm4TlvDq8ikWAM",
    model_id=model,
    output_format="mp3_44100_128",
)
audio_bytes = b"".join(chunk for chunk in audio_stream)
filename = f"../data/voices/output_{model}.mp3"
with open(filename, "wb") as f:
    f.write(audio_bytes)

print("Saved:", filename)

play(audio_bytes)
```

```python
prompt = """
<emotion category="surprised">Ohh is that so ...</emotion>
<emotion category="confident"><style vocal="giggle">You're wrong on this one ...</style></emotion>
<emotion category="excited">Congrats, you've been shortlisted for the next round ...</emotion>
<emotion category="uncertain">Are you sure about that ? ...</emotion>
<emotion category="neutral">Tell me something about yourself ...</emotion>
<emotion category="cheerful">No worries, let's move onto the next question ...</emotion>
<emotion category="sad">I'm sorry to inform you that you've not been shortlisted ...</emotion>
"""

model_ids = [
    "eleven_flash_v2_5",
    "eleven_flash_v2",
    "eleven_multilingual_v2",
    "eleven_turbo_v2_5",
    "eleven_turbo_v2"
]

for model in model_ids:
    audio_stream = elevenlabs.text_to_speech.convert(
        text=prompt,
        voice_id="21m00Tcm4TlvDq8ikWAM",
        model_id=model,
        output_format="mp3_44100_128",
    )
    audio_bytes = b"".join(chunk for chunk in audio_stream)
    filename = f"output_{model}.mp3"
    with open(filename, "wb") as f:
        f.write(audio_bytes)

    print("Saved:", filename)

    play(audio_bytes)
```
